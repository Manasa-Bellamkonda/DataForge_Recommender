```python
import sys
import os
sys.path.append(os.path.abspath("../src"))

import pandas as pd
import numpy as np
from sklearn.metrics import mean_squared_error
from matrix_factorization import MatrixFactorization

train = pd.read_csv("../data/processed/train.csv")
val = pd.read_csv("../data/processed/validation.csv")
test = pd.read_csv("../data/processed/test.csv")
```

```python
mf_model = MatrixFactorization(
    n_factors=20,
    lr=0.01,
    reg=0.02,
    n_epochs=15
)

mf_model.fit(train)

val_preds = mf_model.predict(val)
val_rmse = np.sqrt(mean_squared_error(val["rating"], val_preds))

print("Validation RMSE (MF):", round(val_rmse, 4))
```

```python
test_preds = mf_model.predict(test)
test_rmse = np.sqrt(mean_squared_error(test["rating"], test_preds))

print("Test RMSE (MF):", round(test_rmse, 4))
```
