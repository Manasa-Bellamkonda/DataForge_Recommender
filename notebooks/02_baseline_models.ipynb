```python
import sys
sys.path.append("../src")

import pandas as pd
from baseline import GlobalMeanModel
from sklearn.metrics import mean_squared_error
import numpy as np

# Load processed splits
train = pd.read_csv("../data/processed/train.csv")
val = pd.read_csv("../data/processed/validation.csv")
test = pd.read_csv("../data/processed/test.csv")

print("Train:", train.shape)
print("Validation:", val.shape)
print("Test:", test.shape)
```

```python
# Train global mean model
model = GlobalMeanModel()
model.fit(train)

# Predict on validation set
val_predictions = model.predict(val)

rmse = np.sqrt(mean_squared_error(val["rating"], val_predictions))

print("Validation RMSE (Global Mean):", round(rmse, 4))
```

```python
from baseline import BiasModel
```

```python
# Train bias model
bias_model = BiasModel()
bias_model.fit(train)

# Predict on validation set
val_predictions_bias = bias_model.predict(val)

rmse_bias = np.sqrt(mean_squared_error(val["rating"], val_predictions_bias))

print("Validation RMSE (Bias Model):", round(rmse_bias, 4))
```
